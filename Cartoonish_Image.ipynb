```python
import cv2
from google.colab import files
import matplotlib.pyplot as plt
import io
import numpy as np
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Function to handle image file upload in Google Colab
def upload_image():
    uploaded = files.upload()  # This opens a file upload dialog in Colab
    for filename in uploaded.keys():
        print(f"User uploaded file: {filename}")
        return filename

# Upload the image using Google Colab's file upload
image_path = upload_image()

# Read the uploaded image
color_image = cv2.imread(image_path)

# Check if the image was read properly
if color_image is None:
    print("Failed to load the image, exiting.")
    exit()

# Convert BGR image to RGB for displaying with matplotlib
color_image_rgb = cv2.cvtColor(color_image, cv2.COLOR_BGR2RGB)

# Ask the user for the style selection
cartoon_style_selection = input("This script currently has 2 styles. Please type 1 or 2: ")

# Process the image based on style selection
if cartoon_style_selection == "1":
    # Cartoonify process for style 1
    cartoon_image_style_1 = cv2.stylization(color_image, sigma_s=150, sigma_r=0.25)
    cartoon_image_style_1_rgb = cv2.cvtColor(cartoon_image_style_1, cv2.COLOR_BGR2RGB)
    plt.imshow(cartoon_image_style_1_rgb)
    plt.title("Cartoon Style 1")
    plt.axis('off')  # Turn off axis
    plt.show()

    # Save the cartoonified image to Google Drive
    save_path = '/content/drive/My Drive/cartoon_style_1_image.jpg'
    cv2.imwrite(save_path, cartoon_image_style_1)  # Saving the image
    print(f"Cartoonified image saved at: {save_path}")

elif cartoon_style_selection == "2":
    # Cartoonify process for style 2
    cartoon_image_style_2 = cv2.stylization(color_image, sigma_s=60, sigma_r=0.5)
    cartoon_image_style_2_rgb = cv2.cvtColor(cartoon_image_style_2, cv2.COLOR_BGR2RGB)
    plt.imshow(cartoon_image_style_2_rgb)
    plt.title("Cartoon Style 2")
    plt.axis('off')  # Turn off axis
    plt.show()

    # Save the cartoonified image to Google Drive
    save_path = '/content/drive/My Drive/cartoon_style_2_image.jpg'
    cv2.imwrite(save_path, cartoon_image_style_2)  # Saving the image
    print(f"Cartoonified image saved at: {save_path}")

else:
    print("Invalid style selection.")
```
